# DATA REDUCTION

```python
# The first 2 rows and the last 8 rows are of no use, so we are droping the rows
fbi.drop(fbi.index[0:2], inplace=True)
fbi.drop(fbi.index[-11:], inplace=True)
fbi.reset_index(inplace=True, drop=True)
fbi.head()
```

# Data cleaning

```python
fbi.iloc[0]
```

```python
cols = []

for i in range(0, 4):
    cols.append(fbi.iloc[0][i])

for i in range(4,23):
    if i%2 == 1:
        cols.append(fbi.iloc[0][i])
    
cols
```

```python
# Updating the column names

fbi.columns = fbi.iloc[0]
fbi.drop(fbi.index[0:2], inplace=True)
fbi.reset_index(inplace=True, drop=True)
fbi.head()
```

```python
# Remove Rate Per columns

cols = fbi.columns[fbi.columns.notnull()]
fbi = fbi[cols].copy()
fbi.head()
```

```python
# Removing Percent-Change rows

for index, row in fbi.iterrows():
    if row['Year'] == 'Percent change':
        fbi.drop(index, inplace=True)
fbi.reset_index(inplace=True, drop=True)
fbi.head()
```

```python
# The first 2 rows are of no use, removing first 2 rows!

fbi.drop(fbi.index[0:2], inplace=True)
fbi.reset_index(inplace=True, drop=True)
fbi.head()
```

```python
# Fill the blank spaces in Area column

for i in range(fbi.shape[0]):
    if str(fbi['Area'][i]) == 'nan':
        fbi['Area'][i] = fbi['Area'][i-1]

fbi.head()
```
